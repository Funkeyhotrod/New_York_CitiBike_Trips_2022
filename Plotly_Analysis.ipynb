## Part 2

```python
import streamlit as st
import pandas as pd
import numpy as np
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import matplotlib.pyplot as plt
from datetime import datetime as dt
from streamlit_keplergl import keplergl_static
```

```python
df = pd.read_csv('new_york_data.csv', index_col = 0)
```

```python
df.drop(columns = {'_merge'}, inplace = True)
```

```python
df = df.sort_values(by = 'date')
```

```python
df.head()
```

```python
# Create a month column 

df['date'] = pd.to_datetime(df['date'], format = '%Y-%m-%d')
df['month'] = df['date'].dt.month
df['month'] = df['month'].astype('int')
```

```python
# Create the season column

df['season'] = [
"winter" if (month == 12 or 1 <= month <= 4)
    else "spring" if (4 < month <= 5)
    else "summer" if (6 <= month <= 9)
    else "fall"
for month in df['month']
    ]
```

```python
df.head()
```

## Part 3

```python
# Making a new dataframe that shows the number of times a station has been used

df['value'] = 1 
df_groupby_bar = df.groupby('start_station_name', as_index=False).agg({'value': 'sum'})
top20 = df_groupby_bar.nlargest(20, 'value')
```

```python
# Making a Bar Chart

fig = go.Figure(go.Bar(x = top20['start_station_name'], y = top20['value'], marker={'color': top20['value'],'colorscale': 'Blues'}))
```

```python
fig.update_layout(
    title = 'Top 20 most popular bike stations in New York',
    xaxis_title = 'Start stations',
    yaxis_title ='Sum of trips',
    width = 1500, height = 1000
)
```

```python
# Save the top 20 stations as a CSV file

top20.to_csv('top20.csv')
```

## Part 4

```python
df.head()
```

```python
# Groupby the count of the trips 

df_group = pd.DataFrame(df.groupby(['date'])['ride_id'].count()).reset_index()
```

```python
# Adding new column

df_group.rename(columns = {'ride_id':'daily_bike_rides'}, inplace = True)
df = df.merge(df_group, on = "date", how = 'outer', indicator = True)
print(df['_merge'].value_counts(dropna = False))
print("Shape of January 1st is", df[df['date'] == '2022-01-01'].shape) # Check 
print("Shape of January 2nd is", df[df['date'] == '2022-01-02'].shape) # Second check 
```

```python
df.drop(columns = {'_merge'}, inplace = True)
```

```python
df.head()
```

```python
# Making a subset only with the columns necessary for the line chart

df_lc = df[['date', 'avgTemp', 'daily_bike_rides']]
```

```python
df_lc = (df_lc.drop_duplicates()).reset_index()
```

```python
df_lc.drop(columns = {'index'}, inplace = True)
```

```python
df_lc.head()
```

```python
df_lc.shape
```

```python
# Making a Line chart

fig = make_subplots(specs = [[{"secondary_y": True}]])
```

```python
fig.add_trace(
go.Scatter(x = df_lc['date'], y = df_lc['daily_bike_rides'], name = 'Daily bike rides',
marker = {'color' : df_lc['daily_bike_rides'], 'color' : 'blue'}),
secondary_y = False
)

fig.add_trace(
go.Scatter(x=df_lc['date'], y = df_lc['avgTemp'], name = 'Daily temperature',
marker = {'color' : df_lc['avgTemp'], 'color' : 'red'}),
secondary_y=True
)
```

```python
# Save the dual-axis line chart data as a CSV file

df_lc.to_csv('dual-axis.csv')
```
